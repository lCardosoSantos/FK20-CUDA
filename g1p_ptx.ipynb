`g1p_ptx_fun(int op, g1p_t* out1, g1p_t* out0, g1p_t const* in1, g1p_t const* in0);`

'''
a   -> accumulator
op1, op2 -> operand
x1, y1, z1, z2, y2, z2 -> fp[6]-sized
t0, t1, t2, t3 -> 64biy
'''


`<Label>` is replaced with

```
    mov.u32 %fun label;
    mov.u32 %ret retLabel;
    brx.idx.uni %fun, fp_fun;
retLabel:
```

# Assemblify

````python
import re
import string

def add(t, b, c):
   a=[f"FP_MOV(b, {b})",
      f"FP_MOV(c, {c})",
      f"<fp_add>",
      f"FP_MOV({t}, a)\n"]
   return a

def sub(t, b, c):
   a=[f"FP_MOV(b, {b})",
      f"FP_MOV(c, {c})",
      f"<fp_sub>",
      f"FP_MOV({t}, a)\n" ]
   return a

def mul(t, b, c):
   a=[f"FP_MOV(b, {b})",
      f"FP_MOV(c, {c})",
      f"<fp_mul>",
      f"FP_MOV({t}, a) \n" ]
   return a

def sqr(t, b):
   a=[f"FP_MOV(b, {b})",
      f"<fp_sqr>",
      f"FP_MOV({t}, a)\n" ]
   return a

def x2(t, b):
   a=[f"FP_MOV(a, {b})",
      f"<fp_x2>",
      f"FP_MOV({t}, a)\n" ]
   return a

def x3(t, b):
   a=[f"FP_MOV(a, {b})",
      f"<fp_x3>",
      f"FP_MOV({t}, a)\n" ]
   return a

def x8(t, b):
   a=[f"FP_MOV(a, {b})",
      f"<fp_x8>",
      f"FP_MOV({t}, a)\n" ]
   return a

def x12(t, b):
   a=[f"FP_MOV(a, {b})",
      f"<fp_x12>",
      f"FP_MOV({t}, a)\n" ]
   return a

def cpy(t, b):
   a=[f"FP_MOV({t}, {b})\n"]
   return a

def neg(t, b):
   a=[f"FP_MOV(b, {b})",
      f"<fp_neg>",
      f"FP_MOV({t}, a)\n" ]
   return a

def mma(t, a, b, c, d):
   a=[f"FP_MOV(b, {a})", #op1 = a
      f"FP_MOV(c, {b})", #op2 = b
      f"<fp_mul>",
      f"FP_MOV(b, {c})", #op1 = c
      f"FP_MOV(c, {d})", #op2 = d
      f"FP_MOV({d}, a)", #d = resut
      f"<fp_mul>",
      f"FP_MOV(b, {d})", #op1 = d 
      f"FP_MOV({d}, c)", #d = op2
      f"FP_MOV(c, a)",   #op2 = result
      f"<fp_add>",
      f"FP_MOV({t}, a)\n", #t = result
       ]
   return a
````

````python
dc = {
    "X1":"x",
    "Y1":"y",
    "Z1":"z",
    "X2":"u",
    "Y2":"v",
    "Z2":"w"
}

def assemble(s:string) -> string:
    out = ''
    if s.strip()[:2]=='//':
        out += s + '\n'
        return out
    pars = []
    out += '\t// ' + s + '\n'
    match = re.search(r'\((.*?)\)', s)
    if match:
        extracted_text = match.group(1)  # Get the text inside the parentheses
        pars = [x.strip() for x in extracted_text.split(',')]  # Split using comma and space    
    else:
        print("NO MATCH!")
    
    for i in range(len(pars)):
        if pars[i] in dc.keys():
            pars[i] = dc[pars[i]]

    if 'fp_add' in s:
        for i in add(*pars): out += "\t" + i + '\n'
    if 'fp_sub' in s:
        for i in sub(*pars): out += "\t" + i + '\n'
    if 'fp_mul' in s:
        for i in mul(*pars): out += "\t" + i + '\n'
    if 'fp_sqr' in s:
        for i in sqr(*pars): out += "\t" + i + '\n'
    if 'fp_x2' in s:
        for i in x2(*pars):  out += "\t" + i + '\n'
    if 'fp_x3' in s:
        for i in x3(*pars):  out += "\t" + i + '\n'
    if 'fp_x8' in s:
        for i in x8(*pars):  out += "\t" + i + '\n'
    if 'fp_x12' in s:
        for i in x12(*pars): out += "\t" + i + '\n'
    if 'fp_cpy' in s:
        for i in cpy(*pars): out += "\t" + i + '\n'
    if 'fp_neg' in s:
        for i in neg(*pars): out += "\t" + i + '\n'
    if 'fp_mma' in s:
        for i in mma(*pars): out += "\t" + i + '\n'

    return out
````

````python
def preprocess1(cu, symbol):
    cuPTX = ''
    block = 0
    cuPTX += (f"{symbol}:\n")
    cuPTX += ("\t<LOAD2>\n")
    for line in cu.split('\n'):
        if line.strip() == '':
            cuPTX += (f"// BLOCK {block}\n")
            block+=1
        else:
            cuPTX += assemble(line)
    cuPTX += ("\t<STORE1>")
    cuPTX += ("\n\tret.uni;")
    return cuPTX

````

# Fixing function calls

The notation `<label>` is replaced with
```
    mov.u32 %fun label;
    mov.u32 %ret retLabel;
    brx.idx.uni %fun, fp_fun;
retLabel:
```

The return label can be then in the format "ret"+"parentFunctionName"+"counter". This guarantees uniquenes

````python
def genJumps(asm):
    lines = asm.split("\n")
    out = ''
    jCount = 0
    retLabel = []
    #get fun name
    fname = lines[0]
    match = re.search(r'^(.*?):', fname)
    if match:
        fname = match.group(1).strip()
    else:
        print("no function name found")

    for line in lines:
            match = re.search(r'<([^>]+)>', line)
            if not match:
                out += line + '\n'
            else:
                out += "    // cal " + line.strip()[1:-1] + '\n'
                jTarget = match.group(1).strip()
                retLabel.append(f"ret{fname}_{jCount}")
                jCount+=1
                
                out += f"    mov.u32 %fun, {jTarget}_;\n"
                out += f"    mov.u32 %ret, {retLabel[-1]}_;\n"
                out +=  "    brx.idx.uni %fun, fp_fun;\n"
                out += f"{retLabel[-1]}:\n"

    return out, retLabel


#e, ret = genJumps(example)
````
