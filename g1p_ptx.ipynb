Prototype:

`g1p_ptx_fun(int op, g1p_t* out1, g1p_t* out0, g1p_t const* in1, g1p_t const* in0);`

Registers: 
'''
a   -> accumulator
op1, op2 -> operand
x1, y1, z1, z2, y2, z2 -> fp[6]-sized
t0, t1, t2, t3 -> 64biy
'''


`<Label>` is replaced with

```
    move.u32 %fun label;
    move.u32 %ret retLabel;
    brx.idx.uni %fun, fp_fun;
retLabel:
```

Probably cannot make a C Macro with it, unless we can use `__COUNTER__`

# Assemblify

````python
import re
import string

def add(t, b, c):
   a=[f"FP_MOV(b, {b});",
      f"FP_MOV(c, {c});",
      f"<fp_add>",
      f"FP_MOV({t}, a);"]
   return a

def sub(t, b, c):
   a=[f"FP_MOV(b, {b});",
      f"FP_MOV(c, {c});",
      f"<fp_sub>",
      f"FP_MOV({t}, a);" ]
   return a

def mul(t, b, c):
   a=[f"FP_MOV(b, {b});",
      f"FP_MOV(c, {c});",
      f"<fp_mul>",
      f"FP_MOV({t}, a); " ]
   return a

def sqr(t, b):
   a=[f"FP_MOV(a, {b});",
      f"<fp_sqr>",
      f"FP_MOV({t}, a);" ]
   return a

def x2(t, b):
   a=[f"FP_MOV(a, {b});",
      f"<fp_x2>",
      f"FP_MOV({t}, a);" ]
   return a

def x3(t, b):
   a=[f"FP_MOV(a, {b});",
      f"<fp_x3>",
      f"FP_MOV({t}, a);" ]
   return a

def x8(t, b):
   a=[f"FP_MOV(a, {b});",
      f"<fp_x8>",
      f"FP_MOV({t}, a);" ]
   return a

def x12(t, b):
   a=[f"FP_MOV(a, {b});",
      f"<fp_x12>",
      f"FP_MOV({t}, a);" ]
   return a

def cpy(t, b):
   a=[f"FP_MOV({t}, {b});",]
   return a

def neg(t, b):
   a=[f"FP_MOV(a, {b});",
      f"<fp_neg>",
      f"FP_MOV({t}, a);" ]
   return a

def mma(t, a, b, c, d):
   a=[f"FP_MOV(b, {a});", #op1 = a
      f"FP_MOV(c, {b});", #op2 = b
      f"<fp_mul>",
      f"FP_MOV(b, {c});", #op1 = c
      f"FP_MOV(c, {d});", #op2 = d
      f"FP_MOV({d}, a);", #d = resut
      f"<fp_mul>",
      f"FP_MOV(b, {d});", #op1 = d 
      f"FP_MOV({d}, c);", #d = op2
      f"FP_MOV(c, a);",   #op2 = result
      f"<fp_add>",
      f"FP_MOV({t}, a);", #t = result
       ]
   return a
````

````python
dc = {
    "X1":"x",
    "Y1":"y",
    "Z1":"z",
    "X2":"u",
    "Y2":"v",
    "Z2":"w"
}

def assemble(s:string) -> string:
    if s.strip()[:2]=='//':
        print(s)
        return
    pars = []
    print('\t//', s)
    match = re.search(r'\((.*?)\)', s)
    if match:
        extracted_text = match.group(1)  # Get the text inside the parentheses
        pars = [x.strip() for x in extracted_text.split(',')]  # Split using comma and space    
    
    for i in range(len(pars)):
        if pars[i] in dc.keys():
            pars[i] = dc[pars[i]]

    if 'fp_add' in s:
        for i in add(*pars): print("\t", i, sep='')
    if 'fp_sub' in s:
        for i in sub(*pars): print("\t", i, sep='')
    if 'fp_mul' in s:
        for i in mul(*pars): print("\t", i, sep='')
    if 'fp_sqr' in s:
        for i in sqr(*pars): print("\t", i, sep='')
    if 'fp_x2' in s:
        for i in x2(*pars): print("\t", i, sep='')
    if 'fp_x3' in s:
        for i in x3(*pars): print("\t", i, sep='')
    if 'fp_x8' in s:
        for i in x8(*pars): print("\t", i, sep='')
    if 'fp_x12' in s:
        for i in x12(*pars): print("\t", i, sep='')
    if 'fp_cpy' in s:
        for i in cpy(*pars): print("\t", i, sep='')
    if 'fp_neg' in s:
        for i in neg(*pars): print("\t", i, sep='')
    if 'fp_mma' in s:
        for i in mma(*pars): print("\t", i, sep='')


    print()
````

````python
cu_add = """    
    fp_add(t0, X1, Y1); // t3
    fp_add(t1, Y1, Z1); // t8
    fp_add(t2, Z1, X1); // td

    fp_mul(X1, X1, X2); // t0
    fp_mul(Y1, Y1, Y2); // t1
    fp_mul(Z1, Z1, Z2); // t2

    fp_add(t3, X2, Y2); // t4
    fp_add(Y2, Y2, Z2); // t9
    fp_add(Z2, Z2, X2); // te

    fp_mul(X2, t3, t0); // t5
    fp_mul(Y2, Y2, t1); // ta
    fp_mul(Z2, Z2, t2); // tf

    fp_x3(t0, X1);      // ti
    fp_add(t1, Y1, Z1); // tb
    fp_add(t2, Z1, X1); // tg
    fp_x12(t3, Z1);     // tk

    fp_add(X1, X1, Y1); // t6
    fp_add(Z1, Y1, t3); // tl
    fp_sub(Y1, Y1, t3); // tm

    fp_sub(X1, X2, X1); // t7
    fp_mul(X2, X1, t0); // ts

    fp_mul(X1, X1, Y1); // tp
    fp_mul(Y1, Y1, Z1); // tr

    fp_sub(Y2, Y2, t1); // tc
    fp_mul(Z1, Z1, Y2); // tt
    fp_sub(Z2, Z2, t2); // th

    fp_x12(Z2, Z2);     // tn
    fp_mul(Y2, Y2, Z2); // to
    fp_mul(Z2, Z2, t0); // tq

    fp_sub(X1, X1, Y2); // X3
    fp_add(Y1, Y1, Z2); // Y3
    fp_add(Z1, Z1, X2); // Z3"""
````

````python
block = 0
print("g1p_add:")
print("\t<LOAD2>")
for line in cu_add.split('\n'):
    if line.strip() == '':
        print(f"// BLOCK {block}")
        block+=1
    else:
        assemble(line)
print("\t<STORE1>")
````
